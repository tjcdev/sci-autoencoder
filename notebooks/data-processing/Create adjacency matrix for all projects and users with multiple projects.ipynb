### Add/Remove columns

```python
project_ids = list(projects['project_id'])
count = 0
# Add any missing columns
for project_id in project_ids:
    if project_id not in profilesProjects.columns:
        profilesProjects[project_id] = 0

print(profilesProjects.shape)
        
# Remove any columns that aren't in project_data
cols = profilesProjects.columns.tolist()
for col in cols:
    if col not in list(projects['project_id']):
        profilesProjects = profilesProjects.drop(columns=[col])
    
print(profilesProjects.shape)
```
